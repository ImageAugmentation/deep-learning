```python
from IPython.display import HTML
from base64 import b64encode

def show(filenames):
    html = ''
    if not isinstance(filenames, list):
        filenames = [filenames]
        
    for filename in filenames:
        image = open(filename,'rb').read()
        src = 'data:image/svg+xml;base64,' + b64encode(image).decode()
        html += '<img src="%s" width=480 />' % src 
    return HTML(html)

def play(filename):
    html = ''
    video = open(filename,'rb').read()
    src = 'data:video/mp4;base64,' + b64encode(video).decode()
    html += '<video width=1080 controls autoplay loop><source src="%s" type="video/mp4"></video>' % src 
    return HTML(html)
```

```python
show(["d_fake_loss.svg","d_real_loss.svg","d_loss.svg","g_loss.svg",'d_fake_loss.svg'])
```
